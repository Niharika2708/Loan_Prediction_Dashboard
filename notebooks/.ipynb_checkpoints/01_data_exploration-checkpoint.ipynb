```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv("~/Projects/Python/Loan_Default_Prediction/Data/Loan_Default.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# Replacing null cells with median value in numeric columns

num_cols = ['rate_of_interest', 'Interest_rate_spread', 'Upfront_charges', 'property_value', 'income', 'LTV', 'dtir1']
df.fillna({col: df[col].median() for col in num_cols}, inplace=True)
```

```python
# Replacing null cells with mode values in categorical columns

cat_cols = ['loan_limit', 'approv_in_adv', 'loan_purpose', 'term', 'Neg_ammortization', 'age', 'submission_of_application']
df.fillna({col: df[col].mode()[0] for col in cat_cols}, inplace=True)
```

```python
# Checking for class imbalance in Target Variable: 'Status'

plt.figure(figsize=(6,4))
sns.countplot(x=df['Status'])
plt.title("Loan Default Status Distribution")
plt.show()
```

```python
# Checking feature 1 Distribution

plt.figure(figsize=(10,5))
sns.histplot(df['loan_amount'], bins=30, kde=True)
plt.title("Loan Amount Distribution")
plt.show()
```

```python
# Checking feature 2 Distribution

plt.figure(figsize=(10, 5))
sns.histplot(df['income'], bins=30, kde=True)
plt.title("Income Distribution")
plt.show()
```

```python
# Feature 1 vs Default Relationship

plt.figure(figsize=(8,5))
sns.boxplot(x=df['Status'], y=df['loan_amount'])
plt.title("Loan Amount vs Default")
plt.show()
```

```python
# Feature 2 vs Default Relationship

plt.figure(figsize=(8,5))
sns.boxplot(x=df['Status'], y=df['income'])
plt.title("Income vs Default")
plt.show()
```

```python
# Correlation Matrix 

numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(12,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap = 'coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
# One-hot encoding for categorical variables

df = pd.get_dummies(df, drop_first=True)
```

```python
df.to_csv("~/Projects/Python/Loan_Default_Prediction/Data/cleaned_loan_data.csv", index=False)
```
